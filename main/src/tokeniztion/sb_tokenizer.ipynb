```python
import requests
from sb_tokenizer import SBTokenizer
```

```python
def get_data():
    resp = requests.get("https://www.gutenberg.org/files/35/35-0.txt")
    return resp.text

txt = get_data()
```

```python
actual_text = txt[533:]
print(actual_text[:200])
```

```python
t = SBTokenizer()
t.train(actual_text, vocab_size=10000)
```

```python
len(t.merge_ranks)
```

```python
import string
import re

print("Txt Length ", len(actual_text))

ec = t.encode(actual_text)
print("Encoded tokens", len(ec))
split = re.split(fr"[{string.punctuation}\s]", actual_text)
print("Words = ", len(split))
print("Tokenizerd ", len(t.pretokenize(actual_text)))
```
